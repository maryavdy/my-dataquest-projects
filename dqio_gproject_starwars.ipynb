```python
import pandas as pd
```

```python
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(5)
```

```python
filter_respID_isnull = star_wars["RespondentID"].notnull()
star_wars = star_wars[filter_respID_isnull]
star_wars.head(3)
```

```python
to_bool = {
    "Yes": True,
    "No": False
}

star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(to_bool)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(to_bool)
star_wars.head(5)
```

```python
print(star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts(dropna=False))
print(star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts(dropna=False))
```

```python
import numpy as np

name_to_bool = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(name_to_bool)
```

```python
rename_cols = {
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_ep1",
    "Unnamed: 4": "seen_ep2",
    "Unnamed: 5": "seen_ep3",
    "Unnamed: 6": "seen_ep4",
    "Unnamed: 7": "seen_ep5",
    "Unnamed: 8": "seen_ep6"
}

star_wars = star_wars.rename(columns=rename_cols)
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
rename_cols_ranking = {
    "first_place": "rank_ep1",
    "ranking_2": "rank_ep2",
    "ranking_3": "rank_ep3",
    "ranking_4": "rank_ep4",
    "ranking_5": "rank_ep5",
    "ranking_6": "rank_ep6"
}
star_wars = star_wars.rename(columns=rename_cols_ranking)
star_wars.head()
```

```python
rank_cols = star_wars.columns[9:15]
star_wars[rank_cols].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.bar(range(6), star_wars[rank_cols].mean())
plt.show()
```

**The highest ranked Star Wars movie is Episode 3 at an average score of 4.3 across 835 responses.**

```python
seen_cols = ["seen_ep1", "seen_ep2", "seen_ep3", "seen_ep4", "seen_ep5", "seen_ep6"]
star_wars[seen_cols].sum()
```

```python
plt.bar(range(6), star_wars[seen_cols].sum())
plt.show()
```

**The movie that most people have seen is Episode 5** and yet its ranking is only 2.5 suggesting that not many people enjoyed Episode 5.

## Splitting dataset by binary groups

### Males v Females

```python
import numpy as np

male_fans = star_wars[star_wars["Gender"] == "Male"]
female_fans = star_wars[star_wars["Gender"] == "Female"]
male_fans[rank_cols].mean()
```

```python
# data to plot
n_groups = 6
means_males = (male_fans[rank_cols].mean())
means_females = (female_fans[rank_cols].mean())

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
 
rects1 = plt.bar(index, means_males, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Males')
 
rects2 = plt.bar(index + bar_width, means_females, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Females')

plt.xlabel('Males vs Females')
plt.ylabel('Avg Movie Rankings')
plt.title('Average Star Wars Episode Rankings')
plt.xticks(index + bar_width, ("ep1", "ep2", "ep3", "ep4", "ep5", "ep6"))
plt.legend()
plt.show()
```

```python
# data to plot
n_groups = 6
seen_males = (male_fans[seen_cols].sum())
seen_females = (female_fans[seen_cols].sum())

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
 
rects1 = plt.bar(index, seen_males, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Males')
 
rects2 = plt.bar(index + bar_width, seen_females, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Females')

plt.xlabel('Males vs Females')
plt.ylabel('Audience reach')
plt.title('Audience reach b')
plt.xticks(index + bar_width, ("ep1", "ep2", "ep3", "ep4", "ep5", "ep6"))
plt.legend()
plt.show()
```
